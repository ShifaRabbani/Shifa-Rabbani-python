# Decorators


Timing function execution

```python
import time
def timer(func):
    # arguments and key_word arguments
    def wrapper(*args , **kwargs):
        start = time.time()
        result= func(*args , **kwargs)
        end = time.time()
        print(f"{func.__name__} ran in {end-start} time" )
        return result
    return wrapper

@timer
def example_function(n):
    time.sleep(n)
example_function(2)
```
