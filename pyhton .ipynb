3. Tasks : Bitwise Operation in Details


Bitwise operators work directly on the binary (0 and 1) form of numbers to perform operations bit by bit.

```python
a = 5
b = 3
# and
print(a & b) 
# Becuse = 0101 & 0011 = 0001 → 1
# or
print( a | b)
# Becuse = 0101 | 0011 = 0111 → 7
# xor
print(a ^ b)
# Becude = 0101 ^ 0011 = 0110 → 6
# Not
print(~a)
# Becuse = ~0101 = -(0101 + 1) → -6
```
